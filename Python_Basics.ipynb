# **Data Types**

```python
# Integer
print(10)
# 1*e^308
print(1e308)
```

```python
# Floats/Decimals
print(9.09)
# 1.7*e^308
print(1.7e308)
print(10.11+2.34)
```

```python
# Booleans
print(True)
print(False)
```

```python
# Complex
print(1+2j)
```

```python
# List -> similar to an arrat
print([1,2,3,4,5])
print(['a','b','c'])
print(['x','y','z',1.2,23,True])
```

```python
# Tuples
print((1,2,3,4,5))
print(('a','b','c'))
```

```python
# Sets
print({1,2,3,4,5})
print({'a','b','c'})
```

```python
# Dictionary - looks like a JSON
print({'name':'anurag','age':38, 'city':'delhi'})
```
